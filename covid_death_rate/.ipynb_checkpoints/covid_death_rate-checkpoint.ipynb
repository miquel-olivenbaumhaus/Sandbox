# True death rate

```python
deaths_all = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
infections_all = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
deaths_all =(
                    deaths_all
                    .groupby("Country/Region").sum()
                    .drop(["Lat", "Long"], axis = 1)
                  )


infections_all =(
                        infections_all
                        .groupby("Country/Region").sum()
                        .drop(["Lat", "Long"], axis = 1)
                      )
```

**Let us consider an asymptomatic rate of 80%, which means that 80% of the people infected with the virus, do not experience any symptom**

```python
symptomatic_rate = 0.2
```

```python
death_rate_all = symptomatic_rate*(
                deaths_all
                .div(infections_all)
                .fillna(0)
                .T
)
```

```python
fig, ax = plt.subplots(2, 1)
(100*death_rate_all.iloc[-1]).plot(kind = "bar", ax = ax[0])
sns.kdeplot(100*death_rate_all.iloc[-1], ax = ax[1])

ax[0].set_ylabel("Death Rate[%]")
ax[1].set_xlabel("Death Rate[%]")

fig.tight_layout()
```

## Selected Countries only

```python
selected_countries = ['Japan', 'Singapore', 'Canada',
       'Australia', 'Germany', 'Finland',
       'Sweden','Belgium','Switzerland', 'Austria', 'Israel', 'Norway',  'Andorra', 'US', 'China',
       'Korea, South', 'France', 'Denmark', 'United Kingdom', 'Netherlands']
```

```python
deaths = deaths_all[deaths_all.index.isin(selected_countries)].T
infections =infections_all[infections_all.index.isin(selected_countries)].T.div(symptomatic_rate)

death_rate = deaths.div(infections).fillna(0)
```

```python
fig, ax = plt.subplots(1, 2)
(100*death_rate.iloc[-1]).plot(kind = "bar", ax = ax[0])
sns.distplot(100*death_rate.iloc[-1], hist=False, rug=True);

ax[0].set_ylabel("Death Rate[%]")
ax[1].set_xlabel("Death Rate[%]")
```

```python
(death_rate.iloc[-1][death_rate.iloc[-1] < 0.01].mean())
```

```python
100*(death_rate.iloc[-1][death_rate.iloc[-1] > 0.01].mean())
```

We observe a bimodal distribution for the death rate, with one rate around 0.3% and another around 1.2%.

```python
100*(death_rate.iloc[-1].mean())
```

## Find death rate that minimizes MSE

$$
\mbox{deaths} = \beta\cdot\mbox{infections} +\alpha
$$



Where $\beta$ corresponds to the death rate.

```python
X = infections.iloc[-1].values.reshape(-1, 1)
y = deaths.iloc[-1].values.reshape(-1, 1)
```

```python
reg = LinearRegression(fit_intercept=False).fit(X, y)
```

```python
reg.coef_
```

```python
print("The computed death rate that should be closer on average to the true one is %s%%" %str(round(100*reg.coef_[0][0],1)))
```

# True infections in Spain and Italy

```python
infections_spain_italy = pd.DataFrame(columns = ["Spain", "Italy"], index = ["True", "Confirmed"])

avg_death_rate = reg.coef_[0][0]
```

```python
infections_spain_italy["Spain"] = [deaths_all.T["Spain"].iloc[-1]/avg_death_rate, infections_all.T["Spain"].iloc[-1]]
infections_spain_italy["Italy"] = [deaths_all.T["Italy"].iloc[-1]/avg_death_rate, infections_all.T["Italy"].iloc[-1]]
```

```python
infections_spain_italy
```

```python
fig, ax = plt.subplots()
infections_spain_italy.T.plot(kind = "bar", ax = ax)
ax.set_ylabel("Infections")
ax.set_title("True vs. Confirmed Infections")

#ax.text(0, 200000, "Spain: Percentage Increase vs. confirmed = + 140%")
#ax.text(1+0.1, 200000, "Italy: Percentage Increase vs. confirmed = + 220%")
```

## Word of Caution

What can actually happen in Italy is that beyond a certain amount of infected people, the health care system has to let some people die because of a lack of infrastructres and material. This probably explains the fact that, according to this model, Italy has a 2120% more infected people than confirmed.

A way to test for this hypothesis, would be to investigate whether death rate presents a phase transition: beyond a certain number of infected people, death rate skyrockets.
